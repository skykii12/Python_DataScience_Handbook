**Merge and Concat**

Merge is just like join in SQL, and is used to combine tables together based on some key.

```python
df1 = seattle_weather_df[['date','precipitation','temp_max','temp_min','wind']]
df2 = seattle_weather_df[['date','weather']]

display(df1.head())
display(df2.head())
```

```python
df_merge = pd.merge(df1
        ,df2
        ,on=['date'] #left_on, right_on
        ,how='inner') #left, right, inner, outer
display(df_merge.head())
```

Concat is used for stacking dataframes together. (Equivalent to cbind and rbind in R, depending on what axis is specified).

```python
df_concat = pd.concat([df1,df2], axis=1)
display(df_concat.head())
```

**group by, agg, reset_index**

We often need to group a dataset by some column, and calculate aggregate statistic. For example, we might be interested in the mean temperature for each weather type.

```python
grouped_weather = seattle_weather_df.groupby('weather').mean()
grouped_weather
```

```python
grouped_weather = seattle_weather_df.groupby('weather').agg({
    'temp_max':'mean'
    ,'temp_min':'mean'
})

display(grouped_weather)
print(grouped_weather.columns)
```

You will notice that the weather column's values are in bold. This is because the groupby column has been converted to the index. Sometimes we might want this, but generally I find that it would be more useful if it was just a column. More often than not, after a groupby, I would immediately reset_index.

```python
grouped_weather = seattle_weather_df.groupby('weather').agg({
    'temp_max':'mean'
    ,'temp_min':'mean'
}).reset_index()

display(grouped_weather)
print(grouped_weather.columns)
```

Using .agg is convenient since you can aggregate many columns at once. You can even calculate multiple aggregate calculations for the same column in one go.

```python
grouped_weather = seattle_weather_df.groupby('weather').agg({
    'temp_max':['mean','max','min']
    ,'temp_min':'mean'
}).reset_index()

display(grouped_weather)
print(grouped_weather.columns)
```

You will notice however that because there are multiple aggregate calculations for the same column, an additional hierachical level has been created. This makes it hard to reference a specific column. A good workaround is to immediately reassign the column names based on the aggregate statistic as below.

```python
grouped_weather.columns = [' '.join(col).strip() for col in grouped_weather.columns.values]
display(grouped_weather)
```

**resampling dates**

When dealing with dates, often we want to get regular snapshots at periodic intervals (e.g. weekly data). We can do this easily with resample. For example, we can derive the average values for each column measured at a weekly frequency. Some things to take note of:

- we need to first set the date column to be the index
- specify the frequency of resamping (e.g. 'W', 'Y', 'D', '2W', 'W2-MON')
- perform and aggregate calculation

```python
weekly_df = seattle_weather_df.set_index('date').resample('2W-MON').mean().reset_index()
weekly_df.head()
```

**pandasql**

Using pandasql, we can write sql queries to extract data. Note that using pandas, we can generally do everything sql can, and pandas is more efficient so pandasql should not be used often.

```python
import pandasql

query = """
SELECT 
    *
FROM
    seattle_weather_df
WHERE
    seattle_weather_df.weather in ('sun','fog');
"""
 
query_output = pandasql.sqldf(query, locals())
query_output.head()
```
